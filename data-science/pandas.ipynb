```python
import numpy as np
```

```python
x=np.linspace(0, 2*np.pi, 100)
```

```python
y=np.sin(x)
```

```python
plt.plot(x, y, label="sin")
plt.title("sinusovka")
plt.xlabel("cas")
plt.ylabel("sin()")
plt.legend(loc="upper right")
plt.show()
```

```python
y1=np.sin(x)
y2=np.cos(x)
```

```python
plt.plot(x, y1, "r--", label="sin")
plt.plot(x, y2, "g.", label="cos")
plt.legend(loc="lower left")
```

```python
plt.fill(x, y1, "r", alpha=0.5)
plt.fill(x, y2, "y", alpha=0.5)
plt.plot(x, y1, "b")
```

```python
# první pole hodnot
vals1 = [10, 15, 20, 12, 14, 8]

# druhé pole hodnot
vals2 = [19, 18,  6, 11,  6, 14]

# počet prvků
N = len(vals1)

# indexy prvků
indexes = np.arange(N)

width = 0.30

# sloupcový graf se dvěma skupinami sloupců
plt.bar(indexes-0.15, vals1, width, color='gray', edgecolor='black', label='CPU#1')
plt.bar(indexes+0.25, vals2, width, color='red', edgecolor='black', label='CPU#2')

plt.plot(indexes, vals1)

# povolení zobrazení mřížky
plt.grid(True)

# přidání legendy
plt.legend(loc="upper right")

# zobrazení grafu
plt.show()
```

```python
# náhodné hodnoty

y = np.random.normal(0, 0.1, 100000)

plt.hist(y, bins=100, range=None, density=True)

# zobrazení grafu
plt.show()
```

```python
fig = plt.figure(1, figsize=(6, 6), dpi=100)
ax = fig.add_axes([0.16, 0.16, 0.68, 0.68])

plt.title("Scripting languages")
ax.title.set_fontsize(30)

# popisky jednotlivých výřezů
labels = ['Perl', 'Python', 'Ruby']

# šířky jednotlivých výřezů
fracs = [90, 150, 20]

# vytvoření koláčového grafu
ax.pie(fracs, labels=labels, autopct='%1.1f%%', shadow=False)

# zobrazení grafu
plt.show()
```
