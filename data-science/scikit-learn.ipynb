```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay

# nacteni datove sady
iris = load_iris()

# X je matice (feature matrix)
X = iris.data

# y je vektor (response vector)
y = iris.target

# rozdlen dat
trainX, testX, trainY, testY = train_test_split(X, y, test_size = 0.9)

# konstrukce klasifikatoru
# (s hyperparametrem)
classifier = LogisticRegression(max_iter=1000)

# trening modelu (se vsemi dostupnymi daty)
classifier.fit(trainX, trainY)

y_pred = classifier.predict(testX)

print(classification_report(testY, y_pred))

class_names = iris.target_names

disp = ConfusionMatrixDisplay.from_estimator(
    classifier,
    testX,
    testY,
    display_labels=class_names,
    cmap=plt.cm.Blues,
    normalize=None,
)

print(disp.confusion_matrix)

plt.show()

disp = ConfusionMatrixDisplay.from_estimator(
    classifier,
    testX,
    testY,
    display_labels=class_names,
    cmap=plt.cm.Blues,
    normalize="true",
)

print(disp.confusion_matrix)
```

```python
# model zalozeny na neuronove siti
from sklearn.neural_network import MLPClassifier


# nacteni datove sady
iris = load_iris()

# konstrukce klasifikatoru
# (s hyperparametrem)
classifier = MLPClassifier(max_iter=5000)

# X je matice (feature matrix)
X = iris.data

# y je vektor (response vector)
y = iris.target

# rozdlen na trnovac a testovac data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# trening modelu
classifier.fit(X_train, y_train)

# oekvan vsledky
expexted_labels = y_test

# vsledky modelu (predikovan vsledky)
predicted_labels = classifier.predict(X_test)

# jak je n model spn?
total = 0
same = 0

# porovnn predikce s oekvnm
for (expected, predicted) in zip(expexted_labels, predicted_labels):
    if expected==predicted:
        same+=1
    total+=1

print(f"total:    {total}")
print(f"same:     {same}")
print(f"accuracy: {100.0*same/total:4.1f}%")

print(f"Features: {classifier.n_features_in_}")
print(f"Layers:   {classifier.n_layers_}")
print(f"Outputs:  {classifier.n_outputs_}")
print("Weights:")

for layer, weights in enumerate(classifier.coefs_):
    print("\t", layer, weights.shape)

print("Biases:")

for layer, biases in enumerate(classifier.intercepts_):
    print("\t", layer, biases.shape)
```

```python
# model zalozeny na neuronove siti
from sklearn.neural_network import MLPRegressor

# X je matice, y je vektor
X = np.zeros( (100, 2) )
y = np.zeros( (100, ))
i = 0
for a in range(1, 11):
    for b in range(1, 11):
        X[i,0] = a
        X[i,1] = b
        y[i] = a * b
        i+=1

# konstrukce modelu
nn = MLPRegressor(max_iter=5000, hidden_layer_sizes=(100, 100))

# trnink modelu
nn.fit(X, y)

print(f"Features: {nn.n_features_in_}")
print(f"Layers:   {nn.n_layers_}")
print(f"Outputs:  {nn.n_outputs_}")
print("Weights:")

for layer, weights in enumerate(nn.coefs_):
    print("\t", layer, weights.shape)

print("Biases:")

for layer, biases in enumerate(nn.intercepts_):
    print("\t", layer, biases.shape)
```
